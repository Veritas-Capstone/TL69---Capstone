```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch, numpy as np

model_name = "FacebookAI/roberta-large-mnli" 
tok = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
def fever_predict(claim, evidence_sents):
    premise = " ".join(evidence_sents)
    hyp = claim
    enc = tok(premise, hyp, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        logits = model(**enc).logits[0].cpu().numpy()
    # MNLI label order is [contradiction, neutral, entailment]
    mapping = {2: "SUPPORTED", 0: "REFUTED", 1: "NOT ENOUGH INFO"}
    return mapping[int(np.argmax(logits))], np.argmax(logits).item()
```

```python
print(fever_predict("Vaccines cause Autism", ["Vaccines have a direct link to autism."]))
```
